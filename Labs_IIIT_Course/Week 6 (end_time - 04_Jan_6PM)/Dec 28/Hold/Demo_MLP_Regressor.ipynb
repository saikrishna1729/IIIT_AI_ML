## Import required packages

```
import pandas as pd
from sklearn.neural_network import MLPRegressor
import numpy as np
from sklearn.metrics import r2_score
from numpy.random import seed, rand
```

### PART A:
### Apply MLP regressor for the Linear equation, $y$ = 2 $x_1$ - 3 $x_2$ + 0.5 $x_3$



```
# Randomly generate x1, x2 and x3 of 20 samples
seed(0)
x1 = rand(20)
x2 = rand(20)
x3 = rand(20)
y = 2 * x1 - 3 * x2 +  0.5 * x3
```

```
# X is array of x1, x2, x3
X = np.array([x1, x2, x3]).T

# Create a dataframe with X
linear_data = pd.DataFrame(X, columns = ['x1', 'x2', 'x3'])
linear_data['y'] = y
linear_data.head()
```

```
# Fit the MLP regressor with Linear Equation data
regr1 = MLPRegressor(random_state=42)
X1 = linear_data.iloc[:,:-1]
Y1 = linear_data.iloc[:, -1]
regr1.fit(X1, Y1)
pred1 = regr1.predict(X1)
print(r2_score(pred1, Y1))
```

### PART B:

### Apply MLP regressor for the polynomial equation, $y$ = $x_1^3$ - $ 3 x_2^2$ - $ 4 x_3$ + 2

```
seed(0)
x1 = rand(20)
x2 = rand(20)
x3 = rand(20)
y = x1**3 - 3 * x2**2 - 4 * x3 + 2
```

```
X = np.array([x1, x2, x3]).T
poly_data1 = pd.DataFrame(X, columns = ['x1', 'x2', 'x3'])
poly_data1['y'] = y
poly_data1.head()
```

```
# Fit the MLP regressor
regr2 = MLPRegressor(random_state=42)
X2 = poly_data1.iloc[:,:-1]
Y2 = poly_data1.iloc[:, -1]
regr2.fit(X2, Y2)
pred2 = regr2.predict(X2)
print(r2_score(pred2, Y2))
```

### PART C:

### Apply MLP regressor for the polynomial equation $y$ = $x_1^4$ + $3 x_2^2- sin(x_3) - 2 $

```
seed(0)
x1 = rand(20)
x2 = rand(20)
x3 = rand(20)
y = x1**4 + 3 * x2**2 - np.sin(x3) - 2
```

```
X = np.array([x1, x2, x3]).T
poly_data2 = pd.DataFrame(X, columns = ['x1', 'x2', 'x3'])
poly_data2['y'] = y
poly_data2.head()
```

```
# Fit the MLP regressor
regr3 = MLPRegressor(random_state=42)
X3 = poly_data2.iloc[:,:-1]
Y3 = poly_data2.iloc[:, -1]
regr3.fit(X3, Y3)
pred3 = regr3.predict(X3)
print(r2_score(pred3, Y3))
```
